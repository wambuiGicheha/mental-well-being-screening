### Data Cleaning

```python
#Handle missing values
#check to see if there are null values in percentage form
data.isnull().sum() / len(data) * 100
```

Because the dataset is relatively large, I decides to drop rows that have missing values 

```python
clean_data = data.dropna()
```

```python
# Verify that there are no more missing values
print(data.isnull().sum())
```

Remove duplicates

```python
# Remove duplicates and assign it back to the dataframe
clean_data = clean_data.drop_duplicates()

```

```python
clean_data.info()
```

having 77,051 rows seemed to small so I opted to fill the missing rows in the body column with "'no content" to presere the data set instead.


```python
# Make sure to initialize 'data_main' with 'data' first
data_main = data.copy()  # This creates a copy of the original DataFrame

# Now you can fill missing values in 'body' column for the new 'data_main'
data_main['body'] = data_main['body'].fillna("No content")

# Check the DataFrame info again to confirm the change
data_main.info() 


```

```python
#check for null values 
# Verify that there are no more missing values
data_main.isnull().sum()
```

```python
# Now you can fill missing values in 'body' column for the new 'data_main'
data_main['num_comments'] = data_main['num_comments'].fillna(0)

# Check the DataFrame info again to confirm the change
data_main.info() 
```

```python
#check for null values 
# Verify that there are no more missing values
data_main.isnull().sum()
```

```python
# Convert 'created_utc' column from UTC epoch time to datetime
data_main['created_utc'] = pd.to_datetime(data_main['created_utc'], unit='s')

# To see the updated DataFrame
print(data_main.head(10))
data_main
```

## Exploratory Data Analysis

Univariate Analysis:
For each variable, analyze its distribution (e.g., using histograms, box plots, or data[column].value_counts()).
Summary statistics (data.describe()) for numerical columns can reveal central tendencies and variability.



```python
#summary statistics 
data_main.describe()
```

```python
#Visualize the frequency distribution of values using histograms and to identify outliers using the box plots- for upvotes and num_comments

# Upvotes histogram and box plot
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
data_main['upvotes'].hist(bins=2)
plt.title('Upvotes Distribution')
plt.xlabel('Upvotes')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
data_main['upvotes'].plot(kind='box')
plt.title('Upvotes Box Plot')
plt.show()

#for num_comments
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
data_main['num_comments'].hist(bins=3)
plt.title('Num Comments Distribution')
plt.xlabel('Num Comments')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
data_main['num_comments'].plot(kind='box')
plt.title('Num Comments Box Plot')
plt.show()
```

```python
sns.set(style="whitegrid")
numeric_columns = ['upvotes', 'num_comments', 'label']

# Set up the subplot grid
fig, axes = plt.subplots(1, len(numeric_columns), figsize=(16, 5))

for i, col in enumerate(numeric_columns):
    sns.histplot(data[col], bins=5, kde=True, ax=axes[i], color='skyblue')
    axes[i].set_title(f'Histogram of {col}')
    axes[i].set_xlabel(col.capitalize())
    axes[i].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
#Frequency of categorical columns
# Frequency of 'subreddit' column
print(data_main['subreddit'].value_counts())

# Frequency of 'label' column
print(data_main['label'].value_counts())
```

```python
# Set style
sns.set(style="whitegrid")

# Plot the counts for 'subreddit' column
plt.figure(figsize=(12, 6))
sns.countplot(data=data_main, y='subreddit', order=data_main['subreddit'].value_counts().index, palette='viridis')
plt.title('Counts of Posts by Subreddit')
plt.xlabel('Count')
plt.ylabel('Subreddit')
plt.show()

# Plot the counts for 'label' column
plt.figure(figsize=(8, 6))
sns.countplot(data=data_main, x='label', palette='coolwarm')
plt.title('Counts by Label')
plt.xlabel('Label')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Non-suicidal', 'Suicidal'])
plt.show()
```

```python
#date time column analysis
# Extract year and month for analysis
data_main['year'] = data_main['created_utc'].dt.year
data_main['month'] = data_main['created_utc'].dt.month

# Frequency by year
print(data_main['year'].value_counts())
print(data_main['month'].value_counts())
```

```python
#Analyze Text Column
# Calculate length of 'body' content
data_main['body_length'] = data_main['body'].apply(len)

# Summary statistics of body length
print(data_main['body_length'].describe())

```

### Bivariate Analysis

```python
#Numerical vs. Numerical (Scatter plot and correlation matrix)
#Scatter Plot - Shows the relationship between two numerical variables, upvotes and num_comments.

# Scatter plot for numerical variables
plt.figure(figsize=(8, 6))
sns.scatterplot(x='upvotes', y='num_comments', data=data_main, alpha=0.5)
plt.title("Scatter Plot of Upvotes vs. Number of Comments")
plt.xlabel("Upvotes")
plt.ylabel("Number of Comments")
plt.show()

# Correlation matrix and heatmap
correlation_matrix = data_main[['upvotes', 'num_comments']].corr()
plt.figure(figsize=(6, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)
plt.title("Correlation Matrix of Numerical Variables")
plt.show()
```

```python
# Numerical vs. Categorical (Box plot and Violin plot)
# Box plot of 'upvotes' by 'label'
plt.figure(figsize=(8, 6))
sns.boxplot(x='label', y='upvotes', data=data_main, palette='pastel')
plt.title("Box Plot of Upvotes by Label")
plt.xlabel("Label")
plt.ylabel("Upvotes")
plt.xticks(ticks=[0, 1], labels=["Non-Suicidal (0)", "Suicidal (1)"])
plt.show()

# Violin plot of 'num_comments' by 'subreddit'
plt.figure(figsize=(10, 6))
sns.violinplot(x='subreddit', y='num_comments', data=data_main, palette='muted')
plt.title("Violin Plot of Number of Comments by Subreddit")
plt.xlabel("Subreddit")
plt.ylabel("Number of Comments")
plt.show()
```

```python
# Crosstab between 'subreddit' and 'label'
crosstab = pd.crosstab(data_main['subreddit'], data_main['label'])
print("Crosstab of Subreddit by Label:")
print(crosstab)

# Bar plot for 'subreddit' and 'label'
plt.figure(figsize=(10, 6))
sns.countplot(data=data_main, x="subreddit", hue="label", palette="Set2")
plt.title("Subreddit Distribution by Label")
plt.xlabel("Subreddit")
plt.ylabel("Count")
plt.legend(title="Label", loc='upper right', labels=["Non-Suicidal (0)", "Suicidal (1)"])
plt.show()
```
